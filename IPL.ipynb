```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv

df = pd.DataFrame()

df = pd.read_csv('D:\Data Science\Coursera\Visual final pro\IPL_last_place_data.csv')

df
```

```python
for j in range(len(df)):
    for i in range(len(df.columns)):
        df[df.columns[i]][j] = '0' if df[df.columns[i]][j] == 'DNP' else df[df.columns[i]][j]
        df[df.columns[i]][j] = '0' if df[df.columns[i]][j] == 'DEF' else df[df.columns[i]][j]
        df[df.columns[i]][j] = '0' if df[df.columns[i]][j] == 'SUS' else df[df.columns[i]][j]
        
        df[df.columns[i]][j] = df[df.columns[i]][j][0] if df[df.columns[i]][j][0].isnumeric() == True else df[df.columns[i]][j]
        
        df[df.columns[i]][j] = int(df[df.columns[i]][j]) if i != 0 else df[df.columns[i]][j]
        
df
```

```python
df.replace(0,np.nan,inplace=True)
df.set_index('Team',inplace=True)
df
```

```python
dft=df.T

dft
```

```python
# Droping teams who have not played more than 2 seasons
dft.drop(['Sunrisers Hyderabad','Rising Pune Supergiant','Pune Warriors India','Kochi Tuskers Kerala',
          'Gujarat Lions','Deccan Chargers'],axis=1,inplace=True)
```

```python
plt.style.use('ggplot')
x=[i for i in range(2008,2020)]
dft.plot(figsize=(20,10),grid=True,linewidth=3.0)
plt.legend(prop={'size':16},ncol=4,shadow=True)
plt.plot(dft['Rajasthan Royals'].fillna(method='ffill'), ls = '--', lw = 2,c='darkcyan', label='_nolegend_',alpha=1)
plt.plot(dft['Chennai Super Kings'].fillna(method='ffill'), ls = '--', lw = 2,c='pink', label='_nolegend_',alpha=1)
plt.gca().invert_yaxis()
plt.title('Ranks of 7 IPL Teams across all seasons\n(2008-2019)',size=18)
plt.yticks(fontsize=14,ha='center')
plt.ylabel('Positions',size=15,color='k')
plt.gca().set_xticklabels(x)
plt.xticks(fontsize=14,ha='center')
plt.xlabel('Years',size=15,color='k')
```
